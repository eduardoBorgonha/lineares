```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

Definição dos sinais em tempo discreto:

```python
#intervalo de tempo discreto
n = np.arange(-5, 15)

# Impulso discreto
impulso = np.zeros_like(n)
impulso[n == 0] = 1

# Gate discreto
n0 = 5  # início da função porta
n1 = 10  # fim da função porta
gate = (np.heaviside(n + 1, 1) - np.heaviside(n - 4, 1))

# Exponencial discreta (base e)
#exponencial = np.exp(-n/10)*np.heaviside(n, 1)
# Exponencial discreta com a base que quiser:
exponencial = 2* np.power(0.9, n+1)*np.heaviside(n, 1)
print(exponencial)
plt.figure(figsize=(12, 3))
plt.subplot(1, 3, 1)
plt.stem(n, impulso)
plt.title('Impulso Discreto')
plt.subplot(1, 3, 2)
plt.stem(n, gate,)
plt.title('Gate Discreto')
plt.subplot(1, 3, 3)
plt.stem(n, exponencial)
plt.title('Exponencial Discreta')
plt.tight_layout()
plt.show()
```

Convolução dos sinais em tempo discreto:

```python
#intervalo de tempo discreto para convolução:
n = np.arange(-5, 30)

x = (np.heaviside(n + 1, 1) - np.heaviside(n - 4, 1))
h = np.power(0.9, n) * np.heaviside(n, 1)

y = np.convolve(x, h)

plt.figure(figsize=(10, 3))
plt.subplot(1, 3, 1)
plt.stem(x)
plt.title('x[n]')
plt.subplot(1, 3, 2)
plt.stem(h)
plt.title('h[n]')
plt.subplot(1, 3, 3)
plt.stem(y)
plt.title('y[n] = x[n]*h[n]')
plt.tight_layout()
plt.show()
```
