## Inclui:
* Diagrama de Polos e Zeros
* Vizualização da resposta ao impulso h(t)

Feito para a auxiliar nos estudos da disciplina de Sistemas Lineares usando as bibliotecas Control, NumPy e MatPlotLib.

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct
import warnings
warnings.filterwarnings('ignore')
```

### Abaixo define-se a função transferência.
* O array _numerador_ define os coeficientes do polinômio do numerador de H(s).
* O array _denominador_ define os coeficientes do polinômio do numerador de H(s).
    * Os coeficientes estão em de grau **decrescente**.
* A variável _k_ define um fator multiplicativo para H(s).
#### Perceba que não há necessidade de normalizar a função transferência.
##### OBS: H(S) deve ser própria, grau do numerador maior ou igual ao do denominador.

```python
k = 10
numerador = [1, 100]
denominador = [1, 2, 100]
H = k * ct.tf(numerador, denominador)

print(H)
```

### Resposta o impulso h(t):

```python
t, y = ct.impulse_response(H)

plt.plot(t, y)
plt.xlabel('Tempo (s)')
plt.ylabel('Resposta ao impulso h(t)')
plt.title('Resposta ao Impulso de H(s)')
plt.grid(True)
plt.show()
```

Esse gráfico representa a transformada inversa de Laplace da função Transferência.

## Diagrama de polos e zeros:

```python
ct.pole_zero_plot(H)
```

Ao analisar esse diagrama podemos concluir se o sistema é estável. Caso a região de convergência seja direita (sinal de entrada causal), se todos os polos estiverem no semi-plano esquerdo, o sistema e estável.

## Diagrama de Bode (Módulo e Fase)

```python
w = np.logspace(-1.5, 4, 250) #ajustando a escala logaritmica
ct.bode(H,w,Hz=True,dB=True,deg=False)
```
